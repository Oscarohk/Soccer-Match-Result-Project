```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read Data

```python
from google.colab import drive
drive.mount('/content/drive')

df1 = pd.read_csv('/content/drive/My Drive/Colab Notebooks/COMP 4211 Group Project/EPL_22_23.csv')
df2 = pd.read_csv('/content/drive/My Drive/Colab Notebooks/COMP 4211 Group Project/EPL_21_22.csv')
df3 = pd.read_csv('/content/drive/My Drive/Colab Notebooks/COMP 4211 Group Project/EPL_20_21.csv')
df4 = pd.read_csv('/content/drive/My Drive/Colab Notebooks/COMP 4211 Group Project/EPL_19_20.csv')

df_o = pd.concat([df1, df2, df3, df4], axis = 0)
df_o
```

```python
df = df_o.loc[:, ['MaxCH','MaxCD','MaxCA','AvgCH','AvgCD','AvgCA','MaxC>2.5','AvgC>2.5','MaxCAHH','AvgCAHH','MaxCAHA','AvgCAHA']]
df
```

# Data Exploration

```python
df.info()
```

```python
df.describe().round(2)
```

```python
corr = df.corr()
plt.figure(figsize = (12,8))
plt.tight_layout()
plt.rcParams.update({'font.size': 12})
sns.heatmap(corr, cmap = 'YlGnBu', annot=True, linecolor='r', linewidths=.5)
plt.show()
```

```python
color = sns.color_palette("CMRmap")

sns.set(rc={'figure.figsize':(30,30)}, palette = color, style = 'white')


for i, column in enumerate(df, 1):
    try:
        plt.subplot(6,2,i)
        sns.histplot(df[column]).set(ylabel = None)
        sns.despine(left=True)
    except:
        pass
```

```python
color = sns.color_palette("CMRmap")

sns.set(palette = color, style = 'white', font_scale = 2)

g = sns.pairplot(df)

for ax in g.axes.flatten():
    ax.xaxis.label.set_rotation(90)
    ax.yaxis.label.set_rotation(0)
    ax.yaxis.labelpad = 100
```

```python
# distribution of the 3 cases(H, A, D)
H = df_o['FTR'].value_counts()[0]
A = df_o['FTR'].value_counts()[1]
D = df_o['FTR'].value_counts()[2]

data = {'Home Win':H, 'Away Win':A, 'Draw':D}
labels = list(data.keys())
values = list(data.values())
plt.figure(figsize = (10, 10))
plt.tight_layout()
plt.bar(labels, values)
plt.title("Distribution of the 3 cases(H, A, D)")
plt.show()
```

```python
# historical win percentage
sum = H+A+D
home_win = H/sum
away_win = A/sum
draw = D/sum

print(home_win)
print(away_win)
print(draw)
```
